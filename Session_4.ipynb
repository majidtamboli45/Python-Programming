## Editing item in a list

```python
L = [1,2,3,4,5]

# Editing With Indexing
L[-1] = 500
print(L)

# Editing with slicing

L[1:4] = [200,300,400]

print(L)
```

## Deleting item from a list

```python
# Del
L = [1,2,3,4,5]
print(L)

# Indexing
del L[-1]
print(L)

# Slicing

del L[1:4]
print(L)
```

```python
# remove
L = [1,2,3,4,5]
L.remove(5)
print(L)
```

```python
# Pop
L = [1,2,3,4,5]
L.pop(0)
print(L)
L.pop()
print(L)
```

```python
# Clear
L = [1,2,3,4,5]
L.clear()
print(L)
```

## List Functions

```python
# Len/Min/Max/sorted

L1 = [42,65,74,21,1,6,98]

print(len(L1))
print(min(L1))
print(max(L1))
print(sorted(L1))
print(sorted(L1,reverse=True))
```

```python
# Count

L = [1,2,4,1,2,1,3,1,8,9,1,7,1]
print(L.count(1))
```

```python
# index
L = [1,2,4,1,2,1,3,1,8,9,1,7,1]
print(len(L))
L.index(7)
```

```python
# Reverse

L1 = [42,65,74,21,1,6,98]
L1.reverse()
print(L1)
```

```python
# Sort (vs Sorted)

L = [2,1,5,7,0]
print(L)
print(sorted(L))
print(L)
L.sort()
print(L)
```

```python
# Copy

L = [42,65,74,21,1,6,98]
print(L)
print(id(L))
L1 = L.copy()
print(L1)
print(id(L1))
```

# List Comprehension 

#### Examples

###### 1] Add 1 to 10 Numbers to a list

```python
L = []

for i in range (1,11):
    L.append(i)
print(L)
```

```python
L = [ i for i in range(1,11)]    # Using List Comprehension
print(L)
```

###### 2] Scalar Multiplication on a vector

```python
v = [2,6,4]
s = -2

[s*i for i in v]
```

###### 3] Add Squares

```python
L = [1,2,3,4,5,6,7,8,9]
[i*i for i in L]
```

###### 4] print all numbers divisible by 5 in the range of 1 to 50 

```python
[ i for i in range(1,51) if i % 5 == 0 ]
```

###### 5] Find the languages which start with letter p

```python
languages = ['java','python','php','c','javascript']

[language for language in languages if language.startswith('p')]
```

###### 6] Nested If with list comprehension

```python
# Nested if with List Comprehension
basket = ['apple', 'guava', 'cherry', 'banana']
my_fruits = ['apple', 'kiwi', 'grapes', 'banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

L= []
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

###### 7] Print a (3,3) matrix using list comprehension -> Nested List comprehension

```python
[[i*j for i in range(1,4)] for j in range(1,4)]
```

###### 8] cartesian products -> List comprehension on 2 lists together

```python
L1 = [1,2,3,4]
L2 = [5,6,7,8]
[i*j for i in L1 for j in L2]
```

```python
# 3)
# Count the frequency of a  particular char in a provided string
# eg. 'hello how are you'  the frquency of h is 2 in string
```

```python
s = str(input("Enter the string :- "))
term = str(input("Enter the term :- "))

counter = 0

for i in s:
    if i == term:
        counter += 1

print("frequency:- ",counter)
```

```python
# 4) Write the program which can remove a particular character from string
```

```python
s = str(input("Enter the string :- "))
term = str(input("Enter the term :- "))

result = ' '
for i in s:
    if i!= term:
        result = result + i

print(result)
```

```python
# 5) write the program that can check whether a give string is palindrome or not 
```

```python
s = str(input("Enter the string :- "))
rev = s[::-1]

if s == rev:
    print("yes palindromic ")
else:
    print("Not Palindromic")
```

```python
s = str(input("Enter the string :- "))
flag = True
for i in range(0,len(s)//2):
    if s[i] != s[len(s) - i - 1]:
        flag = False
        print("Not a Palindrome")
        break

if flag:
    print("palindrome")
```

```python
# 6) Write the program to count the number of words in a string without split
```

```python
s = str(input("Enter the string :- "))
temp = ' '
L = []

for i in s:
    if i != ' ':
        temp = temp + i
    else:
        L.append(temp)
        temp = ' '
L.append(temp)
print(L)
        
```

```python
# 7) write the program to convert the string into the title case without using title()
```

```python
s = str(input("Enter the string :- "))
for i in s.split():
    print(i[0].upper() + i[1:].lower())
    
```

```python
s = str(input("Enter the string :- "))
L = []
for i in s.split():
    L.append(i[0].upper() + i[1:].lower())
print(L)
```

```python
# 8) Write A program that convert an integer to string 
```

```python
number = int(input("Enter the number"))
digits = '0123456789'
result = ' ' 
while number != 0:
    result = digits[number % 10] + result
    number = number //10

    print(result)
    print(type(result))
```

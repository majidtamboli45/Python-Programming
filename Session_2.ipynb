##### 3]Logical Operators

```python
print(1 and 0)

print(1 or 0)

print( not 1)
```

##### 4] Bitwise Operators

```python
print(2 & 3)   #Bitwise AND

print(2 | 3)    # Bitwise OR

print( 2 ^ 3)  # Bitwise XOR

print(~3)    # Bitwise Not

print(4 >> 2) # Bitwise left shift

print(4<<2)  # Bitwise Right Shift
```

## 2. If Else in python

##### 1] Login Problem

```python
# Login Program

# Actual Email = majidtamboli45@gmail.com
# Actual Password = 1234

email = input("Enter Your E-mail")
password = input("Enter Your Password")

if email == 'majidtamboli45@gmail.com' and password == '1234':
    print("You are Valid User Welcome")

elif email == 'majidtamboli45@gmail.com' and password != '1234':
    print("Incorrect Password")
    password = input("Enter password again")
    if password == '1234' :
        print("Welcome Finally")
    else:
        print("You Entered Wrong again")
else:
    print("Not Correct Try Again")
    
```

##### 2] Min Of 3 Numbers

```python
num1 = int(input("Enter First Number"))
num2 = int(input("Enter Second Number"))
num3 = int(input("Enter  Third Number"))

if num1 < num2 and num1 < num3:
    print("Smaller Number is :",num1)

elif num2 < num3 :
    print("Smaller Number is :",num2)

else:
    print("Smaller Number is :",num3)
```

##### 3] Menu Driven Calculator

```python
a = int(input("Enter First Number"))
b = int(input("Enter Second Number"))

op = input("Enter operation to be performed")

if op == '+' :
    print(a + b)

elif op == '-':
    print(a-b)

elif op == '*':
    print(a*b)

else:
    print(a/b)

```

## 4.Loops In Python

#### 1] While Loop 

###### 1] To print table

```python
number = int(input("Enter Your Number"))
i = 1
while i < 11:
    print(number , "*" , i , '=' , number*i)
    i += 1
```

```python
# While Loop With Else

x = 1 

while x < 3:
    print(x)
    x += 1
else:
    print("Limit Crossed")
```

###### 2] Guessing Game

```python
import random
jackpot = random.randint(1,100)

guess = int(input("Enter Your Number"))
Attempts = 1
while guess != jackpot:
    if guess < jackpot :
        print("Nope! Guess Higher")
    else:
        print("Nope! Guess Lower")

    guess = int(input("Enter Your Number"))
    Attempts += 1

else:
    print("Hurray You Guess Correct")
    print("Attempts",Attempts)
```

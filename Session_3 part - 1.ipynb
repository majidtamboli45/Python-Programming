##### Sequence Sum

#### Problem :- 1/1! + 2/2! + 3/3!.....

```python
n = int(input("Enter The Number"))
result = 0
fact = 1
for i in range(1,n+1):
    fact = fact*i
    result = result + i / fact

print(result)
```

```python
n = int(input("Enter The Number of rows"))

for i in range(1,n+1):
    for j in range(1,i+1):
        print("*",end=" ")
    print()
    
```

#### 2] Pattern

#### 1
#### 121
#### 12321
#### 1234321

```python
n = int(input("Enter The Number of rows"))

for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end="")
    for k in range(i-1 , 0 , -1):
        print(k,end="")
    print()
```

##### Prime Number

```python
lower = int(input("Enter The Lower range"))
upper = int(input("Enter The upper range"))

for i in range(lower,upper+1):
    for j in range(2,i):
        if i%j == 0:
            break
    else:
        print(i)
```

#### 2] Continue

```python
for i in range(1,10):
    if i == 5:
        continue
    print(i)
```
